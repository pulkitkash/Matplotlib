```python
# Plotting a simple function
price = [48000,49000,50000,60000,55000,41000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
batsman=pd.read_csv('sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Shamra vs V kohli ka carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# changing color by name or hex color code
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='yellow')

plt.title('Rohit Shamra vs V kohli ka carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# Changing line style and line width
#dashdot, dashed, solid, dotted
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dotted',linewidth=5)

plt.title('Rohit Shamra vs V kohli ka carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# Changing marker at corner or points of line like
# 'D', '>', '<', 'o', '+' 
plt.plot(batsman['index'],batsman['V Kohli'],color='green',marker='D',markersize=10,linewidth=5)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='>',markersize=6,linewidth=3)

plt.title('Rohit Shamra vs V kohli ka carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
 
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linewidth=5,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='pink',linewidth=3,label='rohit')

plt.title('Rohit Shamra vs V kohli ka carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs')

plt.legend()  # For labeling of lines
```

we can trim our graph in x and y axis if the variatin is not visible

```python
#WE do this when we got any outlier value in our graph
price = [48000,49000,50000,60000,55000,41000,450000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# we can show grid in our graph
plt.plot(batsman['index'],batsman['V Kohli'],color='green',marker='D',markersize=10,linewidth=5)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='>',markersize=6,linewidth=3)

plt.title('Rohit Shamra vs V kohli ka carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs')

plt.grid()
```
