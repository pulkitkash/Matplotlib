PIE Chart

For Univariate/Bivariate Analysis can be done on categorical vs numerical

USED FOR to find contribution on a standard scale

```python
data= [23,56,78,32,12]
sub=['eng','maths', 'Sci','eco','dbms']
plt.pie(data,labels=sub)
plt.show()  # used for hiding technical information
```

```python
df=pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.2,0,0,0,0,0.1])  # will give us % contribution " %0.1%% "
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.2,0,0,0,0,0.1],shadow=True)  
plt.show()
```
