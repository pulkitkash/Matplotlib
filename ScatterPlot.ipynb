```python
# plt. scatter
x=np.linspace(-10,10,50)
y=10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

```python
df=pd.read_csv('batter.csv')
df=df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='black',marker='+')
plt.title('Avg and SR comparison of top 50 batsman')
plt.xlabel('Average')
plt.ylabel('Strike_rate')
```

```python
tips = sns.load_dataset('tips')
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

```python
# Scatter plot using plt.plot
# its a faster technique
plt.plot(tips['total_bill'],tips['tip'],'o')
```
